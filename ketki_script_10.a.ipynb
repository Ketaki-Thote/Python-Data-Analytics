```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Importing customers_merged

df_ords_prods_merged_final = pd.read_pickle(os.path.join(path, '02 Data','Prepared data', 'customer_merged.pkl'))
```

```python
df_ords_prods_merged_final.head(5)
```

```python
# Addressing PII data

#Information given in the first_name, last_name and gender columns are known as Personally Identifiable Information (PII). These columns can be dropped from the dataset.
```

```python
# Dropping columns to address the PII data

df_ords_prods_merged_final.drop(['first_name', 'last_name', 'Gender'], axis = 1, inplace = True)
```

```python
df_ords_prods_merged_final.head(5)
```

```python
# Creating a regional segmentation of the data

Region = []

for value in df_ords_prods_merged_final["STATE"]:
    if value in ('Maine', 'New Hampshire', 'Vermont','Massachusetts','Rhode Island', 'Connecticut','New York','Pennsylvania', 'New Jersey'):
        Region.append("Northeast")
    elif value in ('Wisconsin','Michigan','Illinois','Indiana','Ohio','North Dakota','South Dakota','Nebraska','Kansas','Minnesota','Iowa','Missouri'):
        Region.append("Midwest")
    elif value in ('Delaware','Maryland','District of Columbia','Virginia','West Virginia', 'North Carolina','South Carolina', 'Georgia', 'Florida','Kentucky', 'Tennessee', 'Mississippi', 'Alabama', 'Oklahoma','Texas', 'Arkansas',' Louisiana'):
        Region.append("South")
    else:
        Region.append("West")
        
```

```python
Region
```

```python
df_ords_prods_merged_final['Region'] = Region
```

```python

df_ords_prods_merged_final['Region'].value_counts(dropna = False)
```

```python
df_ords_prods_merged_final.head(5)
```

```python
# creating crosstab between region and spending_flag

crosstab_regionxspending_flag = pd.crosstab(df_ords_prods_merged_final['spending_flag'], df_ords_prods_merged_final['Region'], dropna = False)
```
