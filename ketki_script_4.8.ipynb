```python
df_ords_prods_merge['mean_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

Trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

```python
df_ords_prods_merge.loc[df_ords_prods_merge['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
 df_ords_prods_merge['median_orders']= df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_orders'] > 20, 'frequency_flag'] = 'Non-Frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_orders'] > 10) & (df_ords_prods_merge['median_orders'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_orders'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
df_ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
